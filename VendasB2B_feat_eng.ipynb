```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carregando os dados

```python
cat_produto = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBCategoriaProduto.csv') #df_categoria_produto
cliente = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBCliente.csv')
depara_cliente_vendedor = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBClientesVendedores.csv')
geografia = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBGeografia.csv')
itens_venda = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBItensVendas.csv') #df_vendas
produto = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBProduto.csv') #df_categoria_produto
regioes = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBRegioesBrasil.csv')
supervisor = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBSupervisor.csv')
vendas = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBVendas.csv') #df_vendas
vendedor = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/PBVendedor.csv')
```

```python
dfDePara = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/tratado/de_para.csv')
```

## Transformando os dados

###Categoria do produto e produto

```python
#renomear columnas
cat_produto.columns = ['categoria_id', 'categoria_descricao']
cat_produto
```

```python
#renomear colunas
produto.rename(columns={'id': 'produto_id', 'descricao': 'produto_descricao'}, inplace=True)
produto.head()
```

#### df_categoria_produto

```python
#merge dos dois datasets
df_categoria_produto = produto.merge(cat_produto, on='categoria_id')
df_categoria_produto.head()
```

###vendas

```python
vendas.head()
```

```python
#renomear coluna id
vendas.rename(columns={'id': 'vendas_id'}, inplace=True)
```

```python
#Transformar os tipos dos dados para podermos calcular os valores
vendas['nf_desconto'] = vendas['nf_desconto'].str.replace(',', '.', regex=True)
vendas['nf_desconto'] = pd.to_numeric(vendas['nf_desconto'], errors='coerce')

vendas['valor_venda'] = vendas['valor_venda'].str.replace(',', '.', regex=True)
vendas['valor_venda'] = pd.to_numeric(vendas['valor_venda'], errors='coerce')
```

###itens venda

```python
itens_venda.head()
```

```python
#renomear a coluna id
itens_venda.rename(columns={'id': 'itens_venda_id'}, inplace=True)
```

#### df_vendas

```python
#merge do vendas e itens_venda pelo vendas_id
df_vendas = vendas.merge(itens_venda, on='vendas_id')
```

```python
#Conferir quantidade de itens vendidos durante os tres anos.
df_vendas.item_quantidade.sum()
```

```python
df_vendas.rename(columns={'valor_bruto_x': 'valor_liquido',
                          'valor_bruto_y': 'valor_bruto'},
                 inplace=True)
df_vendas.head()
```

```python
#calcular a coluna do valor do desconto item a item
valor_unitario = df_vendas['valor_unitario']
valor_bruto = df_vendas['valor_liquido']
desc_nf = df_vendas['nf_desconto']


df_vendas['valor_desconto'] = round(valor_unitario/valor_bruto * desc_nf ,2)
df_vendas.head()
```

### cliente e geografia

```python
#renomear colunas
cliente.rename(columns={
    'id': 'cliente_id',
    'descricao': 'cliente'},
               inplace=True
               )
cliente.head()
```

```python
#renomear colunas
geografia.rename(columns={
    'id':'geografia_id'
}, inplace=True)
```

### regioes

```python
regioes.head()
```

```python
regioes['Regiao'] = regioes['Regiao'].str.extract(r'(\d+)')
```

### depara_cliente-vendedor

```python
depara_cliente_vendedor.head()
```

#### df_clientes_compilado

```python
#merge entre cliente e geografia
df_cliente = cliente.merge(geografia, on='geografia_id')
df_cliente.head()
```

```python
df_clientes = df_cliente.merge(regioes, left_on='uf', right_on='UF').drop('uf', axis=1)
```

```python
df_clientes_vendedor = df_clientes.merge(depara_cliente_vendedor, on='cliente_id')
df_clientes_vendedor.head()
```

```python
df_cliente_compilado = df_clientes_vendedor.merge(dfDePara, on='vendedor_id')
df_cliente_compilado.head()
```

### Vendedor e supervisor

```python
vendedor.rename(columns={'descricao': 'nome_vendedor'}, inplace=True)
supervisor.rename(columns={'descricao': 'nome_supervisor', 'gerente_descricao': 'nome_gerente'}, inplace=True)
```

### df_vendas e df_clientes_compilado

```python
df_cliente_compilado.columns
```

```python
df_vendas.columns
```

#### df_movimentação vendas

```python
df_movimentacao_vendas = df_vendas.merge(df_cliente_compilado, on='cliente_id')
df_movimentacao_vendas.head()
```

#### df_movimentação

```python
df_movimentacao = df_movimentacao_vendas.merge(df_categoria_produto, on='produto_id')
```
