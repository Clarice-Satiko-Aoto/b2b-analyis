# Carregar arquivos necessarios

```python
#carregar arquivo

# df tratado
df_vendas = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/tratado/tratado.csv',
                 parse_dates=['data'])
#de para
dfDePara = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vendas/data/tratado/de_para.csv')

#Metas 2022 e 2023
df_metas = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/vendas/data/tratado/metas22-23.csv",
                       parse_dates=['data'])
```

# merge dos dataframes

```python
#merge do df_metas e df_vendas
#O how='outer' garante que todas as datas presentes nos
#DataFrames de vendas e metas ser√£o mantidas no resultado do merge.

df_historico = df_vendas.merge(df_metas, on=['data', 'gerente_id', 'categoria_id'], how='left')
df_historico.head()
```
